# Visualizations
Plotting our dataset to get an understanding of what the data looks like

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('petroff10')
```

```python
df = pd.read_parquet('steam_dataset/med_sample.parquet')
df.dtypes
```

```python
df
```

```python
plt.hist(df["weighted_vote_score"], bins=20)

plt.title("Histogram of Weighted Vote Scores")
plt.xlabel("Weighted Vote Score")
plt.ylabel("Frequency")
plt.ticklabel_format(scilimits=(-5, 8))
    
plt.show()
```

```python
df["author_playtime_forever"].plot(kind='hist', logy=True, bins=50, edgecolor="black")

plt.title("Histogram of User Playtimes")
plt.xlabel("User Playtime of Reviewed Game (Minutes)")
plt.ylabel("Frequency (log)")

plt.show()
```

```python
df["author_num_games_owned"].plot(kind='hist', logy=True, bins=50, edgecolor='black')

plt.title("Histogram of Owned Games")
plt.xlabel("Number of Games Owned")
plt.ylabel("Frequency")

plt.show()
```

```python
df["steam_purchase"].value_counts().sort_index().plot(kind='bar')

plt.title("Reviews That Purchased the Game From Steam")
plt.xlabel("Purchased From Steam")
plt.ylabel("Frequency")
plt.xticks([0, 1], ["False", "True"])

plt.show()
```

```python
df["received_for_free"].value_counts().sort_index().plot(kind='bar')

plt.title("Reviews That Received the Game For Free")
plt.xlabel("Purchased For Free")
plt.ylabel("Frequency")
plt.xticks([0, 1], ["False", "True"])

plt.show()
```

```python
df["written_during_early_access"].value_counts().sort_index().plot(kind='bar')

plt.title("Reviews That Were Written in Early Access")
plt.xlabel("Written in Early Access")
plt.ylabel("Frequency")
plt.xticks([0, 1], ["False", "True"])

plt.show()
```

```python
plt.figure(figsize=(14, 8))
df["game"].value_counts().head(50).plot(kind='bar')

plt.title("Top 50 Games by Review Count")
plt.xlabel("Games")
plt.ylabel("Review Count")

plt.show()
```
